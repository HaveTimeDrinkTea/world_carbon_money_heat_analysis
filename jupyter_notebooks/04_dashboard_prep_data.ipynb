# Set Up directories and Import Necessary Libraries

```python
# System and OS related tasks
import sys
import os

# Add the project root to Python path
project_root = os.path.abspath('..')
sys.path.insert(0, project_root)

# path to directories
raw_dir = '../data/raw'
processed_dir = '../data/processed'
```

```python
# Get the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid') # sets a white background with grid lines 
```

# Merging the CO<sub>2</sub>, Population, GDP and Energy Sources Data 

📌 Read in the main CO<sub>2</sub>, populations and GDP per capita, PPP (constant 2021 international $) from data/processed/01e_df_co2_pop_gdp_degc_1990_2024.csv. This csv covers data from 1990 to 2024.

```python
data_columns = [
 "country_code_iso3"
    , "country_name_iso3"
    , "year"
    , "co2_pc"
    , "population"
    , "gdp_constant_ppp_pc"
    , "degc"
]
column_dtypes = {
    "country_code_iso3": str
    , "country_name_iso3": str
    , "year": int
    , "co2_pc" : float
    , "population" : float
    , "gdp_constant_ppp_pc": float
    , "degc": float
    }

file_name = "01e_df_co2_pop_gdp_degc_1990_2024.csv"
df_co2_pop_gdp_degc_1990_2024 = pd.read_csv(
 f"{processed_dir}/{file_name}",
    header=0,
    names=data_columns,
    dtype=column_dtypes
    )
```

```python
df_co2_pop_gdp_degc_1990_2024.head(5)
```

📌 Read in the energy sources data from data/processed/01c_df_owid_src_country_2000_2024.csv. This csv covers data from 2000 to 2024.

```python
data_columns = [
 "country"
    , "country_code"
    , "year"
    , "other_renewables"
    , "bioenergy"
    , "solar"
    , "wind"
    , "hydropower"
    , "nuclear"
    , "oil"
    , "gas"
    , "coal"
    , "country_code_iso3"
]
column_dtypes = {
 "country": str
    , "country_code": str
    , "year": int
    , "other_renewables": float
    , "bioenergy": float
    , "solar": float
    , "wind": float
    , "hydropower": float
    , "nuclear": float
    , "oil": float
    , "gas": float
    , "coal": float
    , "country_code_iso3": str
}

file_name = "01c_df_owid_src_country_2000_2024.csv"
df_owid_src_country_2000_2024 = pd.read_csv(
 f"{processed_dir}/{file_name}",
    header=0,
    names=data_columns,
    dtype=column_dtypes
    )
```

```python
df_owid_src_country_2000_2024.head(5)
```

📌 Merge the energy sources data keeping all the rows in main CO<sub>2</sub>, populations and GDP per capita, PPP (constant 2021 international $).

```python
df_all_data_1990_2024 = df_co2_pop_gdp_degc_1990_2024.merge(
    df_owid_src_country_2000_2024
    , left_on=["country_code_iso3","year"]
    , right_on=["country_code_iso3","year"]
    , how="left"
)

df_all_data_1990_2024
```

```python
columns_keep = [
    'country_code_iso3'
    , 'country_name_iso3'
    , 'year'
    , 'co2_pc'
    , 'population'
    , 'gdp_constant_ppp_pc'
    , 'degc'
    , 'other_renewables'
    , 'bioenergy'
    , 'solar'
    , 'wind'
    , 'hydropower'
    , 'nuclear'
    , 'oil'
    , 'gas'
    , 'coal'
]

df_all_data_1990_2024 = df_all_data_1990_2024[columns_keep].copy()

df_all_data_1990_2024
```

📌 export to csv

```python
df_all_data_1990_2024.to_csv(f'{processed_dir}/04_df_all_data_1990_2024.csv', index=False)
print(f"Exported: {processed_dir}/04_df_all_data_1990_2024.csv")
```
